```python
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
df
```

```python
import numpy as np

# Pre-process 'Peak_Hour': Treat 0.0 as a missing value and impute with mode
# First, standardize '0.0' to np.nan to prepare for mode imputation
#df['Peak_Hour'] = df['Peak_Hour'].replace({0.0: np.nan, '0.0': np.nan})

# Get the mode of the non-NaN Peak_Hour values (True/False)
# .mode()[0] is used to get the actual value, as mode() returns a Series
#peak_hour_mode = df['Peak_Hour'].mode()[0]

# Fill the NaN values with the calculated mode
#df['Peak_Hour'] = df['Peak_Hour'].fillna(peak_hour_mode)

# Finally, ensure the column is of boolean type
#df['Peak_Hour'] = df['Peak_Hour'].astype(bool)

# 1. Handle Categorical Constants (Cancellation Reason)
# We assume missing values mean the order was NOT cancelled.
df['Cancellation_Reason'] = df['Cancellation_Reason'].fillna('None')

# 2. Handle Numerical Imputation (Median)
# Using median for columns with outliers like Age, Distance, and Time
numerical_cols = ['Customer_Age', 'Delivery_Time_Min', 'Distance_km', 'Order_Value']
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())

# 3. Handle Categorical Imputation (Mode)
# Filling with the most frequent value for labels
#categorical_cols = ['Customer_Gender', 'City', 'Area', 'Cuisine_Type', 'Payment_Mode', 'Peak_Hour'] # Peak_Hour is handled separately now
#for col in categorical_cols:
#   df[col] = df[col].fillna(df[col].mode()[0])

    # 3. Handle Categorical Imputation (Mode)
# Defining categorical columns based on your dataset headers
categorical_cols = ['Customer_Gender', 'City', 'Area', 'Cuisine_Type', 'Payment_Mode', 'Peak_Hour']

for col in categorical_cols:
    # Ensure the column exists before imputing
    if col in df.columns:
        # Calculate the mode (most frequent value)
        col_mode = df[col].mode()[0]
        # Fill NaN values with the calculated mode
        df[col] = df[col].fillna(col_mode)

# Verification: Check if any NaNs remain in Peak_Hour
print(f"Remaining NaNs in Peak_Hour: {df['Peak_Hour'].isna().sum()}")

# 4. Handle Discounts and Final Amount Logic
# First, assume missing discount means 0 discount
df['Discount_Applied'] = df['Discount_Applied'].fillna(0)

# 5. Recalculate Final_Amount where missing: Final = Value - Discount
# This is more accurate than just using a median.
df['Final_Amount'] = df['Final_Amount'].fillna(df['Order_Value'] - df['Discount_Applied'])

# If any Final_Amount are still missing (due to missing Order_Value), use median
df['Final_Amount'] = df['Final_Amount'].fillna(df['Final_Amount'].median())

# 6. Handle Ratings
# We can use a neutral rating of 3.0 or the median for missing delivery ratings
df['Delivery_Rating'] = df['Delivery_Rating'].fillna(df['Delivery_Rating'].median())

# 7. Handle Time/Date (Forward Fill)
# If the data is chronological, filling from the previous row is best
df['Order_Date'] = df['Order_Date'].fillna(method='ffill')
df['Order_Time'] = df['Order_Time'].fillna(method='ffill')

# 8. Verify the results
print("Missing values after cleaning:")
print(df.isnull().sum())

```
